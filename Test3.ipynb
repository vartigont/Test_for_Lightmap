```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Lightmap3.csv')
```

```python
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'],unit='s')
df['registration_timestamp'] = pd.to_datetime(df['registration_timestamp'],unit='s')
```

```python
df = df.sort_values(['registration_timestamp', 'event_timestamp'])
df['event_date'] = pd.to_datetime(df['event_timestamp']).dt.date
```

```python
df_g = df.groupby('event_date')['inapp_size'].agg(['sum', 'count'])
df_g = df_g.rename(columns={'sum': 'day_sum', 'count': 'count_tr'})
df_g['cumsum_days'] = df_g["day_sum"].cumsum()
df_g
```

```python
df_ln = df_g.drop(['day_sum', 'count_tr'], axis=1)
df_ln
```

```python
df_ln.to_csv('predict.csv')
```
