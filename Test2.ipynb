```python
from numpy import where
```

```python
df = pd.read_csv('Lightmap2.csv')
```

```python
df
```

```python
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'],unit='s')
df
```

```python
df['day'] = pd.to_datetime(df['event_timestamp']).dt.day
df
```

```python
# C 15 по 18 сентября
```

```python
df1 = df[(df['day'] >= 15) & (df['day'] <= 19)]
df1
```

```python
df1_s = df1.sort_values(['player_id', 'event_timestamp', 'event_name'])
df1_s
```

```python
df1_uns = df1_s.groupby('player_id')['event_name'].value_counts().unstack().fillna(0)
df1_uns
```

```python
A_idx = df1_uns.query('install != 0 & login != 0').index.get_level_values('player_id')
res = df1_uns.query('player_id in @A_idx')
res
```

```python
df1_s = df1_s.query('player_id in @A_idx')
df1_s
```

```python
df1_s['day_diff'] = df1_s.groupby('player_id')['day'].diff()
df1_s
```

```python
df1_s = df1_s.query("day_diff != 0")
df1_s
```

```python
df1_s_ad = df1_s[df1_s.duplicated(subset=["player_id"], keep=False)]
df1_s_ad
```

```python
df1_s_ad = df1_s_ad.query("day_diff != 2 and day_diff != 3 and day_diff != 4")
df1_s_ad
```

```python
df1_s_ad['day_diff_ad'] = df1_s_ad.groupby('player_id')['day'].diff()
df1_s_ad
```

```python
df1_s_ad = df1_s_ad.query("day_diff_ad != 2 and day_diff_ad != 3 and day_diff_ad != 4")
df1_s_ad
```

```python
df1_s_ad = df1_s_ad[df1_s_ad.duplicated(subset=["player_id"], keep=False)]
df1_s_ad
```

```python
log1 = len(df1_s_ad['player_id'].unique())
```

```python
df1_i = df1[((df1['day'] < 19) & (df1['event_name'] == 'install'))]
df1_i
```

```python
inst1 = len(df1_i['player_id'].unique())
```

```python
percent1 = log1 / inst1 * 100
percent1
```
